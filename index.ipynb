## Introduction

## Question 1

What is the difference between a NoSQL and a SQL database?  Which is better?

Write your answer below this line: 
The key difference betweeen NoSQL and SQL is that SQL is a relational database that is vertically scalable, has a predefined schema, not suitable for hierarchical databases and can work with complex queries whereas NoSQL is a distributed database that is horizontally scalable, has a dynamic schema, is best suitable for hierarchical databases but cannot work with complex queries. Which is better depends on the data attempting to be stored.







## Question 2

Describe a situation where a NoSQL database might be a better choice than a relational database, and explain your reasoning.

Write your answer below this line: 
For businesses that have rapid growth or databases with no clear schema definitions, NoSQL is the best choice. If you cannot define a schema for your database, if you find yourself denormalizing data schemas, or if your data requirements and schemas are constantly evolving. With mobile apps, such as chatting service, it may be better to use a NoSQL as different messages have constantly changing sizes and number of messages sent may be changing per chat, so SQL may not be the best fit as it has a predetermined schema for its data.






## Question 3

How does MongoDB work? How is it different from a relational database?

Write your answer below this line: 
MongoDB is a document store database. In these Document Stores, each document contains key-value pairs, with the actual data being stored in as the value. This makes Document Stores incredibly flexible, because each document can be unique. With a Document Store, we can just create a key on the fly for the data. For example, storing chat messages between two parties can be stored in Document Store as each unique message sent will be stored as its own document then compiled in a overall document.








## Question 4

What is fault tolerance, and what does it have to do with Resilient Distributed Datasets?

Write your answer below this line: 
Fault tolerance is the property that enables a system to continue operating properly in the event of the failure of some of its components. Resilient Distributed Datasets use fault tolerance by saving backups of data on multiple servers, in the event that one server with the master data isn't runnign properly, the Resilient Distributed Dataset Database may use the backup from another server to fill in.








## Question 5

What is MapReduce? How is it related to Hadoop?

Write your answer below this line: 
MapReduce splits the input data-set into independent tasks which are processed by the map step. The framework sorts the outputs of the maps, which are then input to the reduce tasks. The framework takes care of scheduling tasks, monitoring them and re-executes the failed tasks.
